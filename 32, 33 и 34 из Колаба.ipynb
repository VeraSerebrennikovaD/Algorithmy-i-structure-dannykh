–î–∞–Ω —Å—Ç–µ–∫ –∏ –¥–≤–∞ —ç–ª–µ–º–µ–Ω—Ç–∞  ùê¥
  –∏  ùêµ
 . –ù–µ–æ–±—Ö–æ–¥–∏–º–æ —É–¥–∞–ª–∏—Ç—å –∏–∑ —Å—Ç–µ–∫–∞ –≤—Å–µ —ç–ª–µ–º–µ–Ω—Ç—ã, –∫–æ—Ç–æ—Ä—ã–µ –Ω–∞—Ö–æ–¥—è—Ç—Å—è –º–µ–∂–¥—É  ùê¥
  –∏  ùêµ
  (–≤–∫–ª—é—á–∞—è —Å–∞–º–∏  ùê¥
  –∏  ùêµ
 ).

```python
class Stack:
    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –∫–ª–∞—Å—Å–∞ Stack
    def __init__(self):
        self.items = []  # –°–æ–∑–¥–∞–Ω–∏–µ –ø—É—Å—Ç–æ–≥–æ —Å–ø–∏—Å–∫–∞ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å—Ç–µ–∫–∞

    # –ú–µ—Ç–æ–¥ –¥–ª—è –¥–æ–±–∞–≤–ª–µ–Ω–∏—è —ç–ª–µ–º–µ–Ω—Ç–∞ –≤ —Å—Ç–µ–∫
    def push(self, item):
        self.items.append(item)  # –î–æ–±–∞–≤–ª—è–µ—Ç —ç–ª–µ–º–µ–Ω—Ç –≤ –∫–æ–Ω–µ—Ü —Å–ø–∏—Å–∫–∞

    # –ú–µ—Ç–æ–¥ –¥–ª—è —É–¥–∞–ª–µ–Ω–∏—è –∏ –≤–æ–∑–≤—Ä–∞—â–µ–Ω–∏—è –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ –∏–∑ —Å—Ç–µ–∫–∞
    def pop(self):
        return self.items.pop()  # –£–¥–∞–ª—è–µ—Ç –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –ø–æ—Å–ª–µ–¥–Ω–∏–π —ç–ª–µ–º–µ–Ω—Ç —Å–ø–∏—Å–∫–∞

    # –ú–µ—Ç–æ–¥ –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ —Å—Ç–µ–∫–∞ –±–µ–∑ –µ–≥–æ —É–¥–∞–ª–µ–Ω–∏—è
    def peek(self):
        # –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –ø–æ—Å–ª–µ–¥–Ω–∏–π —ç–ª–µ–º–µ–Ω—Ç, –µ—Å–ª–∏ —Å—Ç–µ–∫ –Ω–µ –ø—É—Å—Ç–æ–π; –∏–Ω–∞—á–µ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç None
        return self.items[-1] if not self.is_empty() else None

    # –ú–µ—Ç–æ–¥ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏, –ø—É—Å—Ç –ª–∏ —Å—Ç–µ–∫
    def is_empty(self):
        return len(self.items) == 0  # –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç True, –µ—Å–ª–∏ —Å–ø–∏—Å–æ–∫ –ø—É—Å—Ç, –∏–Ω–∞—á–µ False

    # –ú–µ—Ç–æ–¥ –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ —Å—Ç–µ–∫–µ
    def size(self):
        return len(self.items)  # –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ —Å–ø–∏—Å–∫–µ
```

```python
def remove_elements_between(stack, A, B):
    temp_stack = Stack()

    # –£–¥–∞–ª—è–µ–º —ç–ª–µ–º–µ–Ω—Ç—ã –¥–æ —Ç–µ—Ö –ø–æ—Ä, –ø–æ–∫–∞ –Ω–µ –Ω–∞–π–¥–µ–º B
    while not stack.is_empty() and stack.peek() != B:
        temp_stack.push(stack.pop())

    # –ï—Å–ª–∏ –Ω–∞–π–¥–µ–Ω B, —É–¥–∞–ª—è–µ–º –µ–≥–æ
    if not stack.is_empty():
        stack.pop()

    # –ü—Ä–æ–¥–æ–ª–∂–∞–µ–º —É–¥–∞–ª—è—Ç—å —ç–ª–µ–º–µ–Ω—Ç—ã, –ø–æ–∫–∞ –Ω–µ –Ω–∞–π–¥–µ–º A
    while not stack.is_empty() and stack.peek() != A:
        stack.pop()

    # –ï—Å–ª–∏ –Ω–∞–π–¥–µ–Ω A, —É–¥–∞–ª—è–µ–º –∏ –µ–≥–æ —Ç–æ–∂–µ
    if not stack.is_empty():
        stack.pop()

    # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –æ—Å—Ç–∞–≤—à–∏–µ—Å—è —ç–ª–µ–º–µ–Ω—Ç—ã –æ–±—Ä–∞—Ç–Ω–æ –≤ –æ—Å–Ω–æ–≤–Ω–æ–π —Å—Ç–µ–∫ –≤ –ø—Ä–∞–≤–∏–ª—å–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ
    while not temp_stack.is_empty():
        stack.push(temp_stack.pop())

# –ò—Å–ø–æ–ª—å–∑—É–µ–º —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—é —Å—Ç–µ–∫–∞ —á–µ—Ä–µ–∑ —Å–ø–∏—Å–æ–∫
stack = Stack()

# –î–æ–±–∞–≤–ª—è–µ–º —ç–ª–µ–º–µ–Ω—Ç—ã –≤ —Å—Ç–µ–∫
print("–°—Ç–µ–∫ –¥–æ —É–¥–∞–ª–µ–Ω–∏—è —ç–ª–µ–º–µ–Ω—Ç–æ–≤:")
for i in range(1, 11):
    stack.push(i)
    print (11 - i)

# –í—ã–∑–æ–≤–µ–º —Ñ—É–Ω–∫—Ü–∏—é —É–¥–∞–ª–µ–Ω–∏—è —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–µ–∂–¥—É A=3 –∏ B=7
remove_elements_between(stack, 3, 7)

print("–°—Ç–µ–∫ –ø–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–µ–∂–¥—É 3 –∏ 7:")
while not stack.is_empty():
    print(stack.pop())
```

```python
class Queue:
    # –ö–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä –æ—á–µ—Ä–µ–¥–∏
    def __init__(self):
        self.items = []  # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –ø—É—Å—Ç–æ–≥–æ —Å–ø–∏—Å–∫–∞ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –æ—á–µ—Ä–µ–¥–∏

    # –ú–µ—Ç–æ–¥ –¥–ª—è –¥–æ–±–∞–≤–ª–µ–Ω–∏—è —ç–ª–µ–º–µ–Ω—Ç–∞ –≤ –∫–æ–Ω–µ—Ü –æ—á–µ—Ä–µ–¥–∏
    def enqueue(self, item):
        self.items.append(item)  # –î–æ–±–∞–≤–ª—è–µ—Ç —ç–ª–µ–º–µ–Ω—Ç –≤ –∫–æ–Ω–µ—Ü —Å–ø–∏—Å–∫–∞

    # –ú–µ—Ç–æ–¥ –¥–ª—è —É–¥–∞–ª–µ–Ω–∏—è —ç–ª–µ–º–µ–Ω—Ç–∞ –∏–∑ –Ω–∞—á–∞–ª–∞ –æ—á–µ—Ä–µ–¥–∏
    def dequeue(self):
        if not self.is_empty():   # –ü—Ä–æ–≤–µ—Ä–∫–∞, –Ω–µ –ø—É—Å—Ç–∞ –ª–∏ –æ—á–µ—Ä–µ–¥—å
            return self.items.pop(0)  # –£–¥–∞–ª—è–µ—Ç –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –ø–µ—Ä–≤—ã–π —ç–ª–µ–º–µ–Ω—Ç —Å–ø–∏—Å–∫–∞, –µ—Å–ª–∏ –æ—á–µ—Ä–µ–¥—å –Ω–µ –ø—É—Å—Ç–∞
        return None  # –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç None, –µ—Å–ª–∏ –æ—á–µ—Ä–µ–¥—å –ø—É—Å—Ç–∞

    # –ú–µ—Ç–æ–¥ –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è —ç–ª–µ–º–µ–Ω—Ç–∞ –≤ –Ω–∞—á–∞–ª–µ –æ—á–µ—Ä–µ–¥–∏ –±–µ–∑ –µ–≥–æ —É–¥–∞–ª–µ–Ω–∏—è
    def peek(self):
        if not self.is_empty():  # –ü—Ä–æ–≤–µ—Ä–∫–∞, –Ω–µ –ø—É—Å—Ç–∞ –ª–∏ –æ—á–µ—Ä–µ–¥—å
            return self.items[0]  # –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –ø–µ—Ä–≤—ã–π —ç–ª–µ–º–µ–Ω—Ç —Å–ø–∏—Å–∫–∞
        return None  # –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç None, –µ—Å–ª–∏ –æ—á–µ—Ä–µ–¥—å –ø—É—Å—Ç–∞

    # –ú–µ—Ç–æ–¥ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏, –ø—É—Å—Ç–∞ –ª–∏ –æ—á–µ—Ä–µ–¥—å
    def is_empty(self):
        return len(self.items) == 0  # –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç True, –µ—Å–ª–∏ –≤ —Å–ø–∏—Å–∫–µ –Ω–µ—Ç —ç–ª–µ–º–µ–Ω—Ç–æ–≤

    # –ú–µ—Ç–æ–¥ –¥–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –æ—á–µ—Ä–µ–¥–∏
    def size(self):
        return len(self.items)  # –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ —Å–ø–∏—Å–∫–µ
```

```python
class Node:
    # –ö–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä —É–∑–ª–∞, –∏—Å–ø–æ–ª—å–∑—É–µ–º–æ–≥–æ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö –≤ –æ—á–µ—Ä–µ–¥–∏
    def __init__(self, data=None):
        self.data = data  # –î–∞–Ω–Ω—ã–µ, —Ö—Ä–∞–Ω—è—â–∏–µ—Å—è –≤ —É–∑–ª–µ
        self.next = None  # –°—Å—ã–ª–∫–∞ –Ω–∞ —Å–ª–µ–¥—É—é—â–∏–π —É–∑–µ–ª –≤ –æ—á–µ—Ä–µ–¥–∏

class Queue:
    # –ö–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä –æ—á–µ—Ä–µ–¥–∏
    def __init__(self):
        self.head = None  # –°—Å—ã–ª–∫–∞ –Ω–∞ –Ω–∞—á–∞–ª–æ (–≥–æ–ª–æ–≤—É) –æ—á–µ—Ä–µ–¥–∏
        self.tail = None  # –°—Å—ã–ª–∫–∞ –Ω–∞ –∫–æ–Ω–µ—Ü (—Ö–≤–æ—Å—Ç) –æ—á–µ—Ä–µ–¥–∏
        self._size = 0    # –°—á–µ—Ç—á–∏–∫ —Ä–∞–∑–º–µ—Ä–∞ –æ—á–µ—Ä–µ–¥–∏

    # –ú–µ—Ç–æ–¥ –¥–ª—è –¥–æ–±–∞–≤–ª–µ–Ω–∏—è —ç–ª–µ–º–µ–Ω—Ç–∞ –≤ –∫–æ–Ω–µ—Ü –æ—á–µ—Ä–µ–¥–∏
    def enqueue(self, item):
        new_node = Node(item)  # –°–æ–∑–¥–∞–Ω–∏–µ –Ω–æ–≤–æ–≥–æ —É–∑–ª–∞ —Å —ç–ª–µ–º–µ–Ω—Ç–æ–º
        if self.tail:  # –ï—Å–ª–∏ –æ—á–µ—Ä–µ–¥—å –Ω–µ –ø—É—Å—Ç–∞
            self.tail.next = new_node  # –°–≤—è–∑–∞—Ç—å –ø–æ—Å–ª–µ–¥–Ω–∏–π —É–∑–µ–ª —Å –Ω–æ–≤—ã–º —É–∑–ª–æ–º
        self.tail = new_node  # –û–±–Ω–æ–≤–∏—Ç—å —Ö–≤–æ—Å—Ç –æ—á–µ—Ä–µ–¥–∏
        if not self.head:  # –ï—Å–ª–∏ –æ—á–µ—Ä–µ–¥—å –±—ã–ª–∞ –ø—É—Å—Ç–∞
            self.head = new_node  # –ù–æ–≤—ã–π —É–∑–µ–ª —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –≥–æ–ª–æ–≤–æ–π
        self._size += 1  # –£–≤–µ–ª–∏—á–µ–Ω–∏–µ —Ä–∞–∑–º–µ—Ä–∞ –æ—á–µ—Ä–µ–¥–∏

    # –ú–µ—Ç–æ–¥ –¥–ª—è —É–¥–∞–ª–µ–Ω–∏—è —ç–ª–µ–º–µ–Ω—Ç–∞ –∏–∑ –Ω–∞—á–∞–ª–∞ –æ—á–µ—Ä–µ–¥–∏
    def dequeue(self):
        if not self.is_empty():  # –ü—Ä–æ–≤–µ—Ä–∫–∞, –Ω–µ –ø—É—Å—Ç–∞ –ª–∏ –æ—á–µ—Ä–µ–¥—å
            removed_data = self.head.data  # –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö —É–¥–∞–ª—è–µ–º–æ–≥–æ —É–∑–ª–∞
            self.head = self.head.next  # –ü–µ—Ä–µ–º–µ—â–µ–Ω–∏–µ –≥–æ–ª–æ–≤—ã –∫ —Å–ª–µ–¥—É—é—â–µ–º—É —É–∑–ª—É
            if self.head is None:  # –ï—Å–ª–∏ –ø–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è –æ—á–µ—Ä–µ–¥—å –ø—É—Å—Ç–∞
                self.tail = None  # –û—á–∏—Å—Ç–∏—Ç—å —Ö–≤–æ—Å—Ç –æ—á–µ—Ä–µ–¥–∏
            self._size -= 1  # –£–º–µ–Ω—å—à–µ–Ω–∏–µ —Ä–∞–∑–º–µ—Ä–∞ –æ—á–µ—Ä–µ–¥–∏
            return removed_data  # –í–æ–∑–≤—Ä–∞—Ç –¥–∞–Ω–Ω—ã—Ö —É–¥–∞–ª—è–µ–º–æ–≥–æ —É–∑–ª–∞
        return None  # –í–æ–∑–≤—Ä–∞—Ç None, –µ—Å–ª–∏ –æ—á–µ—Ä–µ–¥—å –ø—É—Å—Ç–∞

    # –ú–µ—Ç–æ–¥ –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è —ç–ª–µ–º–µ–Ω—Ç–∞ –Ω–∞ –Ω–∞—á–∞–ª–µ –æ—á–µ—Ä–µ–¥–∏ –±–µ–∑ –µ–≥–æ —É–¥–∞–ª–µ–Ω–∏—è
    def peek(self):
        if not self.is_empty():  # –ü—Ä–æ–≤–µ—Ä–∫–∞, –Ω–µ –ø—É—Å—Ç–∞ –ª–∏ –æ—á–µ—Ä–µ–¥—å
            return self.head.data  # –í–æ–∑–≤—Ä–∞—â–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –ø–µ—Ä–≤–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞
        return None  # –í–æ–∑–≤—Ä–∞—Ç None, –µ—Å–ª–∏ –æ—á–µ—Ä–µ–¥—å –ø—É—Å—Ç–∞

    # –ú–µ—Ç–æ–¥ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏, –ø—É—Å—Ç–∞ –ª–∏ –æ—á–µ—Ä–µ–¥—å
    def is_empty(self):
        return self.head is None  # True, –µ—Å–ª–∏ –≥–æ–ª–æ–≤–∞ –æ—á–µ—Ä–µ–¥–∏ None

    # –ú–µ—Ç–æ–¥ –¥–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ä–∞–∑–º–µ—Ä–∞ –æ—á–µ—Ä–µ–¥–∏
    def size(self):
        return self._size  # –í–æ–∑–≤—Ä–∞—â–µ–Ω–∏–µ —Ç–µ–∫—É—â–µ–≥–æ —Ä–∞–∑–º–µ—Ä–∞ –æ—á–µ—Ä–µ–¥–∏
```

```python
class MinQueue:
    def __init__(self):
        self.queue = Queue()  # –û—Å–Ω–æ–≤–Ω–∞—è –æ—á–µ—Ä–µ–¥—å –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è —ç–ª–µ–º–µ–Ω—Ç–æ–≤
        self.mins = Queue()   # –í—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω–∞—è –æ—á–µ—Ä–µ–¥—å –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –º–∏–Ω–∏–º–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π

    def enqueue(self, item):
        self.queue.enqueue(item)  # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–∞ –≤ –æ—Å–Ω–æ–≤–Ω—É—é –æ—á–µ—Ä–µ–¥—å
        # –£–¥–∞–ª–µ–Ω–∏–µ –≤—Å–µ—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –∏–∑ –æ—á–µ—Ä–µ–¥–∏ –º–∏–Ω–∏–º—É–º–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –±–æ–ª—å—à–µ –¥–æ–±–∞–≤–ª—è–µ–º–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞
        while not self.mins.is_empty() and self.mins.peek() > item:
            self.mins.dequeue()
        self.mins.enqueue(item)  # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–∞ –≤ –æ—á–µ—Ä–µ–¥—å –º–∏–Ω–∏–º—É–º–æ–≤

    def dequeue(self):
        item = self.queue.dequeue()  # –£–¥–∞–ª–µ–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–∞ –∏–∑ –æ—Å–Ω–æ–≤–Ω–æ–π –æ—á–µ—Ä–µ–¥–∏
        # –£–¥–∞–ª–µ–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–∞ –∏–∑ –æ—á–µ—Ä–µ–¥–∏ –º–∏–Ω–∏–º—É–º–æ–≤, –µ—Å–ª–∏ –æ–Ω —Ä–∞–≤–µ–Ω —É–¥–∞–ª—è–µ–º–æ–º—É —ç–ª–µ–º–µ–Ω—Ç—É –∏–∑ –æ—Å–Ω–æ–≤–Ω–æ–π –æ—á–µ—Ä–µ–¥–∏
        if item == self.mins.peek():
            self.mins.dequeue()
        return item  # –í–æ–∑–≤—Ä–∞—Ç —É–¥–∞–ª–µ–Ω–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞

    def peek(self):
        return self.queue.peek()  # –ü—Ä–æ—Å–º–æ—Ç—Ä –ø–µ—Ä–≤–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ –≤ –æ—Å–Ω–æ–≤–Ω–æ–π –æ—á–µ—Ä–µ–¥–∏ –±–µ–∑ –µ–≥–æ —É–¥–∞–ª–µ–Ω–∏—è

    def is_empty(self):
        return self.queue.is_empty()  # –ü—Ä–æ–≤–µ—Ä–∫–∞, –ø—É—Å—Ç–∞ –ª–∏ –æ—Å–Ω–æ–≤–Ω–∞—è –æ—á–µ—Ä–µ–¥—å

    def size(self):
        return self.queue.size()  # –ü–æ–ª—É—á–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –æ—Å–Ω–æ–≤–Ω–æ–π –æ—á–µ—Ä–µ–¥–∏

    def get_min(self):
        if not self.mins.is_empty():
            return self.mins.peek()  # –ü–æ–ª—É—á–µ–Ω–∏–µ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞
        return None  # –í–æ–∑–≤—Ä–∞—Ç None, –µ—Å–ª–∏ –æ—á–µ—Ä–µ–¥—å –º–∏–Ω–∏–º—É–º–æ–≤ –ø—É—Å—Ç–∞

# –°–æ–∑–¥–∞–µ–º —ç–∫–∑–µ–º–ø–ª—è—Ä MinQueue
q = MinQueue()

# –î–æ–±–∞–≤–ª–µ–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –æ—á–µ—Ä–µ–¥—å
q.enqueue(5)
print("–î–æ–±–∞–≤–ª–µ–Ω–Ω–æ –≤ –∫–æ–Ω–µ—Ü –æ—á–µ—Ä–µ–¥–∏:", 5, "–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ:", q.get_min())
q.enqueue(3)
print("–î–æ–±–∞–≤–ª–µ–Ω–Ω–æ –≤ –∫–æ–Ω–µ—Ü –æ—á–µ—Ä–µ–¥–∏:", 3, "–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ:", q.get_min())
q.enqueue(7)
print("–î–æ–±–∞–≤–ª–µ–Ω–Ω–æ –≤ –∫–æ–Ω–µ—Ü –æ—á–µ—Ä–µ–¥–∏:", 7, "–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ:", q.get_min())
q.enqueue(1)
print("–î–æ–±–∞–≤–ª–µ–Ω–Ω–æ –≤ –∫–æ–Ω–µ—Ü –æ—á–µ—Ä–µ–¥–∏:", 1, "–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ:", q.get_min())
q.enqueue(9)
print("–î–æ–±–∞–≤–ª–µ–Ω–Ω–æ –≤ –∫–æ–Ω–µ—Ü –æ—á–µ—Ä–µ–¥–∏:", 9, "–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ:", q.get_min())

# –£–¥–∞–ª–µ–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –∏ –ø—Ä–æ–≤–µ—Ä–∫–∞ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è
print("–£–±—Ä–∞–Ω–æ –∏–∑ –Ω–∞—á–∞–ª–∞ –æ—á–µ—Ä–µ–¥–∏:", q.dequeue(), "–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ:", q.get_min())
print("–£–±—Ä–∞–Ω–æ –∏–∑ –Ω–∞—á–∞–ª–∞ –æ—á–µ—Ä–µ–¥–∏:", q.dequeue(), "–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ:", q.get_min())
print("–£–±—Ä–∞–Ω–æ –∏–∑ –Ω–∞—á–∞–ª–∞ –æ—á–µ—Ä–µ–¥–∏:", q.dequeue(), "–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ:", q.get_min())
print("–£–±—Ä–∞–Ω–æ –∏–∑ –Ω–∞—á–∞–ª–∞ –æ—á–µ—Ä–µ–¥–∏:", q.dequeue(), "–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ:", q.get_min())
print("–£–±—Ä–∞–Ω–æ –∏–∑ –Ω–∞—á–∞–ª–∞ –æ—á–µ—Ä–µ–¥–∏:", q.dequeue(), "–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ:", q.get_min())
```
