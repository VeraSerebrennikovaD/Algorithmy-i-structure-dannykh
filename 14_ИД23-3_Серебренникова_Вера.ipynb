10) Написать 4 функции порождения списка целых значений: отсортированного (по возрастанию и убыванию), случайного и почти отсортированного по возрастанию. Каждая функция принимает длину списка в качестве аргумента.

<div class="alert-info">
Стандартный способ

```python
import random

# Функция для списка, отсортированного по возрастанию
def sorted_list(length):
    # Возвращает список от 1 до заданной длины по возрастанию.
    return list(range(1, length + 1))

# Функция для списка, отсортированного по убыванию
def reversed_sorted_list(length):
    # Возвращает список от заданной длины до 1 по убыванию.
    return list(range(length, 0, -1))

# Функция для случайного списка
def random_list(length):
    # Возвращает список случайных чисел заданной длины от 1 до 100.
    return [random.randint(1, 100) for i in range(length)]

# Функция дляпочти отсортированного списка по возрастанию
def almost_sorted_list(length):
    '''Возвращает практически отсортированный список от 1 до заданной длины по возрастанию,
    путем замены одного элемента случайным числом от 1 до 100.'''
    lst = list(range(1, length + 1))
    random_index = random.randint(0, length - 1)
    lst[random_index] = random.randint(1, 100)
    return lst

# Пример использования функций
print("Отсортированный по возрастанию список:", sorted_list(10))
print("Отсортированный по убыванию список:", reversed_sorted_list(10))
print("Случайный список:", random_list(10))
print("Почти отсортированный список по возрастанию:", almost_sorted_list(10))
```

Другой способ работы nearly_sorted_ascending (почти отсортированный список). Считает, выбирая 10% случайных индексов

```python
# Функция для генерации почти отсортированного списка по возрастанию
def nearly_sorted_ascending(length):
    base_list = list(range(1, length+1))
    random_indices = random.sample(range(length), length//10)  # Выбираем 10% случайных индексов
    # Функция sample в модуле random используется для выборки уникальных элементов из последовательности заданного размера.
    random_values = [random.randint(1,100) for i in range(length//10)]  # Генерируем случайные значения
    for i in range(length//10):
        base_list[random_indices[i]] = random_values[i]  # Заменяем 10% элементов случайными значениями
    return base_list


# Пример работы
print("Почти отсортированный список по возрастанию:", nearly_sorted_ascending(10))
```

11) Реализовать функцию тестирования алгоритма сортировки. Процедура принимает на вход алгоритм сортировки и алгоритм порождения списка f_gen(list_len), генерирующий список заданной длины, количество повтороений теста и список длин использующихся для тестирования. Для каждой длины списка функция тестирования возвращает минимальное, среднее и наибольшее количество сравнений и перестановок.

```python
import statistics

# Функция для тестирования алгоритма сортировки
def test_sorting_algorithm(sort_func, list_generator, repetitions, lengths):
    results = {}  # Создаем пустой словарь для результатов
    for length in lengths:  # Проходим по всем длинам списков для тестирования
        comparisons = []  # Список для хранения количества сравнений
        swaps = []  # Список для хранения количества перестановок
        for i in range(repetitions):  # Повторяем тест определенное количество раз
            # Генерируем список с помощью переданной функции list_generator
            lst = list_generator(length)
            # Считаем количество сравнений и перестановок для отсортированного списка
            comparisons_count, swaps_count = sort_func(lst) 
            comparisons.append(comparisons_count)  # Добавляем количество сравнений в список
            swaps.append(swaps_count)  # Добавляем количество перестановок в список
        # Записываем результаты теста в словарь
        results[f'Для длины списка {length}'] = {
            'Минимальное количество сравнений': min(comparisons),  
            'Среднее количество сравнений': statistics.mean(comparisons), 
            'Максимальное количество сравнений': max(comparisons), 
            'Минимальное количество перестановок': min(swaps),  
            'Среднее количество перестановок': statistics.mean(swaps),  
            'Максимальное количество перестановок': max(swaps) 
        }
    return results  # Возвращаем словарь с результатами теста

# Функция для сортировки пузырьком для проверки работы test_sorting_algorithm
def bubble_sort(lst):
    comparisons = 0  # Инициализируем счетчик сравнений
    swaps = 0  # Инициализируем счетчик перестановок
    n = len(lst)  # Длина списка
    for i in range(n):  # Проходим по всем элементам списка
        for j in range(n - i - 1):  # Проходим по элементам до n-i-1
            comparisons += 1  # Увеличиваем счетчик сравнений
            if lst[j] > lst[j + 1]:  # Если текущий элемент больше следующего
                lst[j], lst[j + 1] = lst[j + 1], lst[j]  # Меняем их местами
                swaps += 1  # Увеличиваем счетчик перестановок
    return comparisons, swaps  # Возвращаем количество сравнений и перестановок

# Тестирование алгоритма "пузырек" 
print("Тестирование алгоритма пузырек:")
print("=" * 50) # Просто для красивого вывода
print("Сортировка отсортированного списка:")
list_lengths = [10, 100, 1000]  # пример списка длин для тестирования
repetitions = 10  # количество повторений теста
results = test_sorting_algorithm(bubble_sort, sorted_list, repetitions, list_lengths)
print(results)
```

15) Реализовать модификацию сортировки выбором на основе выбора из подсписков (описанного в лекции). Протестировать алгоритм.



```python
def modified_selection_sort(arr):
    sublist_size = 3  # Размер подсписка (например, 3, как здесь, может быть произвольным)
    comparisons = 0  # Переменная для подсчета количества сравнений
    swaps = 0  # Переменная для подсчета количества перестановок
    n = len(arr)  # Длина списка

    for i in range(0, n, sublist_size):  # Проходим по списку с шагом равным размеру подсписка
        end = min(i + sublist_size, n)  # Определение конца подсписка. Это чтобы не выйти за края списка 
        for j in range(i, end):  # Проходим по элементам подсписка до предпоследнего
            min_index = j  # Инициализируем минимальное значение индекса
            for k in range(j + 1, end - 1):  # Проходим по элементам подсписка от следующего элемента до конца 
                comparisons += 1  # Увеличиваем счетчик сравнений
                if arr[k] < arr[min_index]:  # Если текущий элемент меньше минимального
                    min_index = k  # Обновляем минимальный индекс
            if min_index != j:  # Если найденный минимальный элемент не находится на текущей позиции
                arr[j], arr[min_index] = arr[min_index], arr[j]  # Меняем текущий элемент с минимальным
                swaps += 1  # Увеличиваем счетчик перестановок

    return comparisons, swaps


# Этот алгоритм делит список на подсписки и находит минимальный и максимальный элементы в каждом подсписке. Затем он обменивает минимальный элемент с первым элементом подсписка, а максимальный элемент с последним элементом подсписка.
print("Тестирование алгоритма сортировки выбором на основе выбора из подсписков :")
print("=" * 50)
results = test_sorting_algorithm(modified_selection_sort, random_list, repetitions, list_lengths)
print(results)
```

16) Реализовать быструю сортировку с возможностью подсчета операций. Сравнить производительность алгоритма на случайных и упорядоченных (и почти упорядоченных) данных.



```python
# Быстрая сортировка - это алгоритм сортировки, который выбирает последний элемент из списка (называемый опорным (pivot)),
# затем перераспределяет остальные элементы списка так, чтобы элементы меньше опорного находились слева от него, 
# а элементы больше опорного - справа. Затем процесс рекурсивно повторяется для обеих частей списка, 
# пока вся последовательность не будет отсортирована.

def quick_sort(arr):
    comparisons = 0
    swaps = 0

    stack = [(0, len(arr) - 1)]

    while stack:
        low, high = stack.pop()
        pivot_index = random.randint(low, high)
        pivot = arr[pivot_index]
        arr[pivot_index], arr[high] = arr[high], arr[pivot_index]  # перемещаем опорный элемент в конец
        i = low - 1
        for j in range(low, high):
            comparisons += 1
            if arr[j] < pivot:
                i += 1
                arr[i], arr[j] = arr[j], arr[i]
                swaps += 1
        arr[i + 1], arr[high] = arr[high], arr[i + 1]
        swaps += 1

        # Добавляем новые интервалы для сортировки в стек
        if low < i:
            stack.append((low, i))
        if i + 2 < high:
            stack.append((i + 2, high))

    return comparisons, swaps

import time 

# Функция для замера времени выполнения сортировки
def measure_sorting_time(sort_algorithm, arr):
    start_time = time.time()  # Получаем время начала выполнения. Как бы ставим точку на временной линии, 
    # с которой начиначем засекать время
    sort_algorithm(arr)  # Выполняем алгоритм сортировки
    end_time = time.time()  # Получаем время завершения выполнения. Останавливаем таймер
    return end_time - start_time  # Возвращаем затраченное время

# Сравнение производительности на случайных данных
print("Сравнение производительности на случайных данных:")
print("=" * 50)
for length in list_lengths:
    lst = random_list(length)  # Генерируем случайный список заданной длины
    time_taken = measure_sorting_time(quick_sort, lst.copy())  # Измеряем время выполнения алгоритма быстрой сортировки
    # lst.copy(), чтобы не изменять исходный список. Тз этого не требует, но было бы странно изменять изначальный список
    print(f"Длина списка: {length}, Время выполнения: {time_taken:.6f} сек")  # Выводим результат
print("=" * 50)

# Сравнение производительности на упорядоченных данных
print("Сравнение производительности на упорядоченных данных:")
print("=" * 50)
for length in list_lengths:
    lst = sorted_list(length)  # Генерируем упорядоченный список заданной длины
    time_taken = measure_sorting_time(quick_sort, lst.copy())  # Измеряем время выполнения алгоритма быстрой сортировки
    print(f"Длина списка: {length}, Время выполнения: {time_taken:.6f} сек")  # Выводим результат
    # time_taken:.6f означает вывод секунд с точностью до 6 знаков после запятой (например, 0.000001 сек)
print("=" * 50)

# Сравнение производительности на почти упорядоченных данных
print("Сравнение производительности на почти упорядоченных данных:")
print("=" * 50)
for length in list_lengths:
    lst = almost_sorted_list(length)  # Генерируем почти упорядоченный список заданной длины
    time_taken = measure_sorting_time(quick_sort, lst.copy())  # Измеряем время выполнения алгоритма быстрой сортировки
    print(f"Длина списка: {length}, Время выполнения: {time_taken:.6f} сек")  # Выводим результат
print("=" * 50)
```

17) Улучшить работу быстрой сортировки на упорядоченных и почти упорядоченных за счет изменения алгоритма выбора элемента для разделения массива.

```python
def quick_sort(arr):
    comparisons = 0
    swaps = 0

    stack = [(0, len(arr) - 1)]

    while stack:
        low, high = stack.pop()
        pivot_index = random.randint(low, high)
        pivot = arr[pivot_index]
        arr[pivot_index], arr[high] = arr[high], arr[pivot_index]  # перемещаем опорный элемент в конец
        i = low - 1
        for j in range(low, high):
            comparisons += 1
            if arr[j] < pivot:
                i += 1
                arr[i], arr[j] = arr[j], arr[i]
                swaps += 1
        arr[i + 1], arr[high] = arr[high], arr[i + 1]
        swaps += 1

        # Добавляем новые интервалы для сортировки в стек
        if low < i:
            stack.append((low, i))
        if i + 2 < high:
            stack.append((i + 2, high))

    return comparisons, swaps
#Меняем последний эллемент на рандомный. Это может помочь избежать худшего случая быстрой сортировки на упорядоченных или почти упорядоченных данных, так как вероятность выбора "плохого" опорного элемента снижается.
# Сравнение производительности на случайных данных
print("Сравнение производительности на случайных данных:")
print("=" * 50)
for length in list_lengths:
    lst = random_list(length)
    time_taken = measure_sorting_time(quick_sort, lst.copy())
    print(f"Длина списка: {length}, Время выполнения: {time_taken:.6f} сек")
print("=" * 50)

# Сравнение производительности на упорядоченных данных
print("Сравнение производительности на упорядоченных данных:")
print("=" * 50)
for length in list_lengths:
    lst = sorted_list(length)
    time_taken = measure_sorting_time(quick_sort, lst.copy())
    print(f"Длина списка: {length}, Время выполнения: {time_taken:.6f} сек")
print("=" * 50)

# Сравнение производительности на почти упорядоченных данных.
print("Сравнение производительности на почти упорядоченных данных:")
print("=" * 50)
for length in list_lengths:
    lst = almost_sorted_list(length)
    time_taken = measure_sorting_time(quick_sort, lst.copy())
    print(f"Длина списка: {length}, Время выполнения: {time_taken:.6f} сек")
print("=" * 50)
```

18) Реализовать эффективный алгоритм получения из большого количества отсортированных списков одного отсортированного списка.



```python
# 1. Инициализируем пустой список, который будет содержать итоговый отсортированный список.
# 2. Сравниваем первые элементы всех отсортированных списков и выбираем наименьший элемент. Добавляем его в итоговый список.
# 3. Удаляем выбранный элемент из исходного списка, из которого он был выбран.
# 4. Повторяем шаги 2-3, пока все списки не будут пустыми.


def merge_sorted_lists(lists):
    result = []  # Создаем пустой список для хранения отсортированных значений
    
    while True:  # Запускаем бесконечный цикл 
        min_index = None  # min_index хранит индекс списка с минимальным первым элементом
        
        # Ищем циклом список с минимальной переменной:
        for i, lst in enumerate(lists):  # Итерируемся по спискам и их индексам. В i - индексы списков, в lst - списки
            if lst:  # Если список не пуст
                if min_index == None or lst[0] < lists[min_index][0]:  # Если текущий список имеет меньший первый элемент
                    min_index = i  # Обновляем индекс списка, хранящего минимальную переменную
                    
        if min_index == None:  # Если не найден минимальный элемент
            break  # Прерываем цикл. Это произойдет, если все элементы во всех списках "закончатся"
            
        result.append(lists[min_index][0])  # Добавляем минимальный элемент в результат
        lists[min_index] = lists[min_index][1:]  # Удаляем этот элемент из списка, обновляя список. Он имеет 0-ой индекс, так как
        # списки по условию отсортированные
    return result  # Возвращаем отсортированный список


# Пример использования
sorted_lists = [[1, 3, 5, 7],
                [2, 4, 6, 8],
                [0, 9, 10]]

sorted_result = merge_sorted_lists(sorted_lists)  # Объединяем отсортированные списки
print(sorted_result)  # Вывод результата
```

19) Модифицировать последовательность длин шагов в сортировке Шелла. 19) Модифицировать последовательность длин шагов в сортировке Шелла.
На 2[𝑛/2^(𝑘+1)]+1

т.е. при первой сортировке будут использоваться шаги: 2[n/4]+1, 2[n/8]+1, ... , 3, 1 .


```python
'''Сортировка Шелла - это метод сортировки вставками. Она работает путем сравнения элементов, расположенных на определенном 
расстоянии друг от друга. Сначала сортируются элементы, находящиеся на большом расстоянии друг от друга, а затем расстояние 
между элементами уменьшается с каждой итерацией. Это позволяет элементам "быстрее" перемещаться к своим конечным позициям, 
уменьшая количество сдвигов.'''

def shell_sort_modified(lst):
    n = len(lst)  # Длина списка для сортировки
    k = 1  
    b = 0  # b - флаг того, что gap достигло единицы 
    gap = 2 * (n // (2 ** (k + 1))) + 1  # gap - шаг в сортировке Шелла. Формула из условия
    while gap > 0:  # Пока промежуток существует
        for i in range(gap, n):  # Проходим по элементам списка, начиная с gap
            temp = lst[i]  # Сохраняем текущий элемент списка
            j = i  # Запоминаем текущий индекс
            # Сортировка вставками внутри подсписков с шагом gap
            while j >= gap and lst[j - gap] > temp:  # Пока элементы внутри подсписков больше текущего элемента
                lst[j] = lst[j - gap]  # Перемещаем элементы внутри подсписков
                j -= gap  # Двигаемся по подсписку с шагом gap
            lst[j] = temp  # Сохраняем текущий элемент в результирующий список
        k += 1  # Увеличиваем значение k
        gap = 2 * (n // (2 ** (k + 1))) + 1  # Вычисляем новый размер промежутка
        if gap == 1:  # Если размер промежутка равен 1
            if b == 0:  # Если флаг еще не сломан
                b += 1  # Ломаем флаг
            else:  
                break  # Иначе не ломаем
    return lst  # Возвращаем отсортированный список


# Пример использования
print("=" * 50)
primer_list = random_list(10)
print(f'Изначальный список со случайнфми значениями: {primer_list}')
sorted_arr = shell_sort_modified(primer_list)  # Сортировка случайного списка
print(f'Отсортированный список: {sorted_arr}')  # Вывод отсортированного списка
```
